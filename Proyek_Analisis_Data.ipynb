### Cleaning Data
**Pada tahap ini kita akan membersihkan berbagai masalah yang teridentifikasi dalam proses assessing data.**

**Cleaning data merupakan pembersihan data sebelum melanjutkan ke analisis data. Agar datanya bersih dan terstuktur dengan aman sehingga dapat digunakan dalam pemodelan atau visualisasi data.**

```python
day_df.rename(columns={'instant':'Instant',
                       'dteday':'Datetimes',
                       'season':'Season',
                       'yr':'Year',
                       'mnth':'Month',
                       'holiday':'Holiday',
                       'weekday':'Weekday',
                       'workingday':'Workingday',
                       'weathersit':'WeatherSituation',
                       'temp':'Temperature',
                       'atemp':'ApparentTemperature',
                       'hum':'Humidity',
                       'windspeed':'WindSpeed',
                       'casual':'Casual',
                       'registered':'Registered',
                       'cnt':'Count'},inplace=True) 

```

**Karena tidak titemukan masalah pada data. maka kita akan melakukan perubahan ke kolom 'Datetimes' menjadi tipe data datetime. Dan membersihkan data dengan menghapus baris yang memiliki nilai duplikat**

```python
day_df['Datetimes'] = pd.to_datetime(day_df['Datetimes'], errors='coerce')

day_df_cleaned = day_df.drop_duplicates(subset='Datetimes')

print(day_df_cleaned)
```

```python
hour_df['Datetimes'] = pd.to_datetime(hour_df['Datetimes'], errors='coerce')

hour_df_cleaned = hour_df.drop_duplicates(subset='Datetimes')
print(hour_df_cleaned)
```

## Exploratory Data Analysis (EDA)

### Explore Data gunanya untuk menjawab pertanyaan tersebut

```python
day_df.describe(include="all")
```

```python
hour_df.describe(include="all")
```

```python
season_pivot = day_df.pivot_table(values='Count', index='Season', aggfunc=['count', 'mean', 'std', 'min', 'max']).reset_index()
season_pivot.columns = ['Season', 'Count_Count', 'Count_Mean', 'Count_Std', 'Count_Min', 'Count_Max']
print("\nTabel Pivot untuk Eksplorasi Perbedaan Tren Penggunaan Sepeda pada Musim Tertentu:")
print(season_pivot)
```

```python
workingday_pivot = day_df.pivot_table(values='Count', index='Workingday', aggfunc=['count', 'mean', 'std', 'min', 'max']).reset_index()
workingday_pivot.columns = ['Workingday', 'Count_Count', 'Count_Mean', 'Count_Std', 'Count_Min', 'Count_Max']
print("Tabel Pivot untuk Eksplorasi Pola Penggunaan Sepeda pada Hari Kerja dan Akhir Pekan:")
print(workingday_pivot)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:
Bagaimana distribusi jumlah peminjaman sepeda pada setiap bulan selama dua tahun?


```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Count', data=day_df, estimator='sum', ci=None, palette='muted')
plt.title('Distribusi Jumlah Peminjaman Sepeda pada Setiap Bulan')
plt.xlabel('Bulan')
plt.ylabel('Total Jumlah Peminjaman Sepeda')
plt.show()
```

### Pertanyaan 2:
Bagaiamana kondisi cuaca dapat berkaitan dengan penggunaan sepeda?


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='WeatherSituation', y='Count', data=day_df, palette='viridis')
plt.title('Korelasi antara Kondisi Cuaca dan Penggunaan Sepeda')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Peminjaman Sepeda')
plt.show()
```

## Conclusion

Terdapat pola dengan puncak aktivitas, terjadi pada bulan-bulan musim semi dan musim panas pada bulan Maret hingga September. Ini menyimpulkan bahwa orang-orang cenderung menggunakan sepeda saat cuaca lebih hangat.
